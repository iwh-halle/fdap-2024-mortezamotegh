```python
import re
import PyPDF2
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import nltk

# Ensure the necessary NLTK resources are downloaded
nltk.download('stopwords')
nltk.download('wordnet')

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text

def preprocess_text(text, use_stemming=False, use_lemmatization=False):
    # Remove punctuation, numbers, and special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Convert text to lower case
    text = text.lower()
    
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    text_tokens = text.split()
    filtered_text = [word for word in text_tokens if word not in stop_words]
    
    if use_stemming:
        stemmer = PorterStemmer()
        filtered_text = [stemmer.stem(word) for word in filtered_text]
    
    if use_lemmatization:
        lemmatizer = WordNetLemmatizer()
        filtered_text = [lemmatizer.lemmatize(word) for word in filtered_text]
    
    return ' '.join(filtered_text)

# Correct file path
pdf_path = '/workspaces/fdap-2024-mortezamotegh/qcad015.pdf'

# Extract text from PDF
raw_text = extract_text_from_pdf(pdf_path)

# Preprocess the text
preprocessed_text = preprocess_text(raw_text, use_stemming=False, use_lemmatization=True)

print(preprocessed_text[:1000])  # Print the first 1000 characters of the preprocessed text
```

```python
import re
import PyPDF2
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from textblob import TextBlob
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import spacy

# Ensure the necessary NLTK resources are downloaded
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text

def preprocess_text(text):
    # Remove punctuation, numbers, and special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Convert text to lower case
    text = text.lower()
    
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    text_tokens = text.split()
    filtered_text = [word for word in text_tokens if word not in stop_words]
    
    # Lemmatize the text
    lemmatizer = WordNetLemmatizer()
    lemmatized_text = [lemmatizer.lemmatize(word) for word in filtered_text]
    
    return ' '.join(lemmatized_text)

# Extract text from PDF
pdf_path = '/workspaces/fdap-2024-mortezamotegh/qcad015.pdf'
raw_text = extract_text_from_pdf(pdf_path)

# Preprocess the text
preprocessed_text = preprocess_text(raw_text)

# Frequency Analysis
word_counts = Counter(preprocessed_text.split())
common_words = word_counts.most_common(20)
print("Most common words:", common_words)

# Generate Word Cloud
wordcloud = WordCloud(width=800, height=400).generate(preprocessed_text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Sentiment Analysis
blob = TextBlob(preprocessed_text)
sentiment = blob.sentiment
print("Sentiment Analysis:", sentiment)

# Summarization
parser = PlaintextParser.from_string(raw_text, Tokenizer("english"))
summarizer = LsaSummarizer()
summary = summarizer(parser.document, 5)  # Summarize to 5 sentences
summary_text = ' '.join([str(sentence) for sentence in summary])
print("Summary:", summary_text)

# Contextual Analysis - Topic Modeling with LDA using sklearn
vectorizer = CountVectorizer(stop_words='english')
doc_term_matrix = vectorizer.fit_transform([preprocessed_text])
lda = LatentDirichletAllocation(n_components=5, random_state=0)
lda.fit(doc_term_matrix)

# Display the topics
for i, topic in enumerate(lda.components_):
    print(f"Topic {i}:")
    print([vectorizer.get_feature_names_out()[index] for index in topic.argsort()[-10:]])
```
